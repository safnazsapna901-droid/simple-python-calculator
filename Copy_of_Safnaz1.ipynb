```python
# Import required libraries
import matplotlib.pyplot as plt

# Sample data
students = ["Asha", "Ravi", "Kiran", "Meena", "John"]
marks = [85, 78, 92, 69, 88]

# Create the bar chart
plt.bar(students, marks, color='skyblue')

# Add labels and title
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Marks Chart")

# Show the graph
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot sine wave
plt.plot(x, y, label="sin(x)", color="blue")

# Add labels and title
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Sine Wave")
plt.legend()

# Show the graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
fruits = ["Apples", "Bananas", "Cherries", "Mangoes"]
quantities = [25, 30, 15, 20]

# Create pie chart
plt.pie(quantities, labels=fruits, autopct="%1.1f%%", startangle=90)

# Add title
plt.title("Fruit Distribution")

# Show chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
expenses = [1200, 900, 1100, 1500, 1000, 1300]

# Plot line chart
plt.plot(months, expenses, marker='o', color='green')

# Add labels and title
plt.xlabel("Months")
plt.ylabel("Expenses (in ₹)")
plt.title("Monthly Expenses")
plt.grid(True)

# Show chart
plt.show()
```

```python
# Pyramid star pattern

rows = 5
for i in range(1, rows+1):
    print(" " * (rows - i) + "*" * (2*i - 1))
```

```python
# Number triangle pattern

rows = 5
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
# Square star pattern

rows = 5
for i in range(rows):
    for j in range(rows):
        print("*", end=" ")
    print()
```

```python
# Hollow square star pattern

rows = 5
for i in range(rows):
    for j in range(rows):
        if i == 0 or i == rows-1 or j == 0 or j == rows-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
# Right-angled triangle star pattern

rows = 5
for i in range(1, rows+1):
    print("* " * i)
```

```python
# Inverted right-angled triangle star pattern

rows = 5
for i in range(rows, 0, -1):
    print("* " * i)
```

```python
# Diamond star pattern

rows = 5

# Upper part
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (2*i - 1))

# Lower part
for i in range(rows-1, 0, -1):
    print(" " * (rows - i) + "*" * (2*i - 1))
```

```python
# Hollow diamond star pattern

rows = 5

# Upper part
for i in range(1, rows + 1):
    for j in range(1, rows - i + 1):
        print(" ", end="")
    for j in range(1, 2*i):
        if j == 1 or j == 2*i - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()

# Lower part
for i in range(rows - 1, 0, -1):
    for j in range(1, rows - i + 1):
        print(" ", end="")
    for j in range(1, 2*i):
        if j == 1 or j == 2*i - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
# Pascal's Triangle

rows = 6
for i in range(rows):
    print(" " * (rows - i), end="")  # spacing
    num = 1
    for j in range(i + 1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)  # formula for next number
    print()
```

```python
# Floyd's Triangle

rows = 6
num = 1
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(num, end=" ")
        num += 1
    print()
```
